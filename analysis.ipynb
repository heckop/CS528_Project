```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## <b> Fetching the Dataset and Plotting the Heatmap for the task executed by the servers throughout different timeslots

```python
dataset_df = pd.read_csv('data.csv')
dataset_df = dataset_df.dropna()
heatmap = sns.heatmap(dataset_df, annot=True)
plt.show()
```

## <b> Plotting the bar graph for different timeslots for different servers

```python
for i,d in enumerate(dataset_df.columns):
    plt.figure(i)
    sns.displot(dataset_df[d], kde=False, bins=20)
    plt.title("timeSlot: "+d)
    plt.show()
```

## <b> Plotting the sum of executed task on a server throughout all time slots

```python
sum_row=dataset_df.sum(axis=1)
sum_row.plot(kind='bar')
plt.show()
```
